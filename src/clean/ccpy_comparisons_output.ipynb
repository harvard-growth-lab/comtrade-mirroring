```python
df = merged_ccpy[['exporter', 'importer', 'value_final_py', 'value_final_stata']].groupby(['exporter', 'importer']).agg('sum')
df = df[~((df.value_final_py==0.0) & (df.value_final_stata!=0.0))]
df['ccy_difference'] = df['value_final_stata'] - df['value_final_py']
df['perc_diff'] = 100 * ((df['value_final_stata'] - df['value_final_py']) / (df['value_final_stata'] + df['value_final_py']))
df = df.sort_values(by="perc_diff")
```

```python
df["perc_diff"].describe()
```

```python
exp_df = df.reset_index()
exp_df = exp_df[['exporter', 'value_final_py', 'value_final_stata']].groupby('exporter').agg('sum')
# exp_df = exp_df[~((exp_df.value_final_py==0.0) & (exp_df.value_final_stata!=0.0))]
exp_df['cy_difference'] = exp_df['value_final_stata'] - exp_df['value_final_py']
exp_df['perc_diff'] = 100 * ((exp_df['value_final_stata'] - exp_df['value_final_py']) / (exp_df['value_final_stata'] + exp_df['value_final_py']))
exp_df = exp_df.sort_values(by="perc_diff")
                                                                        
```
